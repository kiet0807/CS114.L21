```python
import pandas as pd
```

```python
info = pd.read_csv('./wecode.csv', encoding='utf-8', sep=',')
```

### Tổng số proplem đã solve / tổng số proplem submit

```python
for i in range(1,info["user_id"].max()):
    solved = info.loc[info['user_id'] == i, 'is_final'].sum()
    submit = (info.loc[info['user_id'] == i, 'user_id'].sum())/i
    print("user_id:",i,",solved = ",solved,",submit =", submit)
    if submit != 0:
        print(solved/submit)
    else:
        print(0);
```
